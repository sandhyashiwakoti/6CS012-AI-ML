## Task 1: Unit Converter
Create a Python program that converts between different units of measurement (Length, Weight, Volume). The program prompts the user to choose a conversion type, input a value, performs the conversion, handles errors, and displays the result.

```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'm_to_ft' or 'ft_to_m'
    Returns:
        float: The converted value.
    """
    if direction == 'm_to_ft':
        return value * 3.28084
    elif direction == 'ft_to_m':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'm_to_ft' or 'ft_to_m'.")

def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'kg_to_lbs' or 'lbs_to_kg'
    Returns:
        float: The converted value.
    """
    if direction == 'kg_to_lbs':
        return value * 2.20462
    elif direction == 'lbs_to_kg':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'kg_to_lbs' or 'lbs_to_kg'.")

def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'l_to_gal' or 'gal_to_l'
    Returns:
        float: The converted value.
    """
    if direction == 'l_to_gal':
        return value * 0.264172
    elif direction == 'gal_to_l':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'l_to_gal' or 'gal_to_l'.")

def unit_converter():
    """
    Main function that prompts user to choose conversion type and value.
    Returns:
        float: The converted result.
    """
    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        choice = input("Enter 1, 2, or 3: ")

        if choice == "1":
            print("a. Meters to Feet  b. Feet to Meters")
            sub = input("Enter a or b: ")
            value = float(input("Enter value: "))
            direction = 'm_to_ft' if sub == 'a' else 'ft_to_m'
            result = convert_length(value, direction)
            print(f"Result: {result:.4f}")

        elif choice == "2":
            print("a. Kg to Lbs  b. Lbs to Kg")
            sub = input("Enter a or b: ")
            value = float(input("Enter value: "))
            direction = 'kg_to_lbs' if sub == 'a' else 'lbs_to_kg'
            result = convert_weight(value, direction)
            print(f"Result: {result:.4f}")

        elif choice == "3":
            print("a. Liters to Gallons  b. Gallons to Liters")
            sub = input("Enter a or b: ")
            value = float(input("Enter value: "))
            direction = 'l_to_gal' if sub == 'a' else 'gal_to_l'
            result = convert_volume(value, direction)
            print(f"Result: {result:.4f}")

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
            return None

        return result

    except ValueError as e:
        print(f"Error: Invalid input! Please enter a numeric value. Details: {e}")
        return None

# Run the converter
unit_converter()
```

## Task 2: Mathematical Operations on a List
Create a Python program that performs various mathematical operations (sum, average, maximum, minimum) on a user-inputted list of numbers, with error handling for invalid inputs and empty lists.

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The sum of all numbers in the list.
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The average of all numbers in the list.
    """
    return sum(numbers) / len(numbers)

def calculate_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The maximum value in the list.
    """
    return max(numbers)

def calculate_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The minimum value in the list.
    """
    return min(numbers)

def math_operations():
    """
    Main function that prompts user to choose an operation
    and input a list of numbers, then performs the operation.
    Returns:
        float: The result of the selected operation.
    """
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter 1, 2, 3, or 4: ")
        numbers = input("Enter numbers separated by spaces: ")

        # Convert input string to list of floats
        numbers = [float(x) for x in numbers.split()]

        if len(numbers) == 0:
            raise ValueError("List is empty!")

        if choice == "1":
            result = calculate_sum(numbers)
            print(f"Sum: {result}")
        elif choice == "2":
            result = calculate_average(numbers)
            print(f"Average: {result}")
        elif choice == "3":
            result = calculate_maximum(numbers)
            print(f"Maximum: {result}")
        elif choice == "4":
            result = calculate_minimum(numbers)
            print(f"Minimum: {result}")
        else:
            print("Invalid choice!")
            return None

        return result

    except ValueError as e:
        print(f"Error: Invalid input! {e}")
        return None

# Run the function
math_operations()
```

# 4.2 Exercise on List Manipulation

## 1. Extract Every Other Element:


```python
def extract_every_other(lst):
    """
    Extracts every other element starting from the first.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Every other element starting from index 0.
    """
    return lst[::2]
```

## 2. Slice a Sublist:


```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from start to end index (inclusive).
    Parameters:
        lst (list): Input list.
        start (int): Starting index.
        end (int): Ending index (inclusive).
    Returns:
        list: Sublist from start to end.
    """
    return lst[start:end+1]
```

## 3. Reverse a List Using Slicing:

```python
def reverse_list(lst):
    """
    Reverses a list using slicing.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Reversed list.
    """
    return lst[::-1]
```

## 4. Remove the First and Last Elements:

```python
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.
    Parameters:
        lst (list): Input list.
    Returns:
        list: List without first and last elements.
    """
    return lst[1:-1]
```

## 5. Get the First n Elements:

```python
def get_first_n(lst, n):
    """
    Returns the first n elements of a list.
    Parameters:
        lst (list): Input list.
        n (int): Number of elements to return.
    Returns:
        list: First n elements.
    """
    return lst[:n]
```

## 6. Extract Elements from the End:

```python
def get_last_n(lst, n):
    """
    Returns the last n elements of a list.
    Parameters:
        lst (list): Input list.
        n (int): Number of elements to return.
    Returns:
        list: Last n elements.
    """
    return lst[-n:]
```

## 7. Extract Elements in Reverse Order:

```python
def reverse_skip(lst):
    """
    Returns every second element starting from second-to-last, moving backward.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Every second element in reverse order.
    """
    return lst[-2::-2]
```

## Testing All Functions:

```python
# Testing all functions
lst1 = [1, 2, 3, 4, 5, 6]
lst2 = [1, 2, 3, 4, 5]

print("Original list:", lst1)
print("1. Extract every other:", extract_every_other(lst1))
print("2. Sublist (index 2 to 4):", get_sublist(lst1, 2, 4))
print("3. Reversed:", reverse_list(lst2))
print("4. Remove first and last:", remove_first_last(lst2))
print("5. First 3 elements:", get_first_n(lst2, 3))
print("6. Last 2 elements:", get_last_n(lst2, 2))
print("7. Reverse skip:", reverse_skip(lst1))
```

# 4.3 Exercise on Nested List


## 1. Flatten a Nested List:

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.
    Parameters:
        lst (list): A nested list.
    Returns:
        list: A flattened single-dimension list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result
```

## 2. Accessing Nested List Elements:


```python
def access_nested_element(lst, indices):
    """
    Accesses a specific element in a nested list using indices.
    Parameters:
        lst (list): A nested list.
        indices (list): List of indices to navigate to the element.
    Returns:
        The element at the specified position.
    """
    for index in indices:
        lst = lst[index]
    return lst
```

## 3. Sum of All Elements in a Nested List:


```python
def sum_nested(lst):
    """
    Calculates the sum of all elements in a nested list.
    Parameters:
        lst (list): A nested list.
    Returns:
        float: Sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
```

## 4. Remove Specific Element from a Nested List:

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.
    Parameters:
        lst (list): A nested list.
        elem: The element to remove.
    Returns:
        list: Modified nested list with element removed.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result
```

## 5. Find the Maximum Element in a Nested List:

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list.
    Parameters:
        lst (list): A nested list.
    Returns:
        The maximum element.
    """
    flat = flatten(lst)
    return max(flat)
```

## 6. Count Occurrences of an Element in a Nested List:


```python
def count_occurrences(lst, elem):
    """
    Counts occurrences of a specific element in a nested list.
    Parameters:
        lst (list): A nested list.
        elem: The element to count.
    Returns:
        int: Number of occurrences.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count
```

## 7. Flatten a List of Lists of Lists:


```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list regardless of depth.
    Parameters:
        lst (list): A deeply nested list.
    Returns:
        list: A fully flattened single-dimension list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result
```

## 8. Nested List Average:


```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.
    Parameters:
        lst (list): A nested list.
    Returns:
        float: Average of all elements.
    """
    flat = flatten(lst)
    return sum(flat) / len(flat)
```

## Testing all functions


```python
print("1. Flatten:", flatten([[1, 2], [3, 4], [5]]))
print("2. Access nested element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1,2]))
print("3. Sum nested:", sum_nested([[1, 2], [3, [4, 5]], 6]))
print("4. Remove element:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
print("5. Find max:", find_max([[1, 2], [3, [4, 5]], 6]))
print("6. Count occurrences:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
print("7. Deep flatten:", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
print("8. Average nested:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

### 2. Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1     
x − y + 2z = −3     
3x + y − z = 9

```python
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3,  1, -1]])
B_eq = np.array([-1, -3, 9])

# Using inverse method
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("Solution using Inverse Method:")
print(f"x = {X[0]:.2f}, y = {X[1]:.2f}, z = {X[2]:.2f}")

# Using np.linalg.solve
X_solve = np.linalg.solve(A_eq, B_eq)
print("\nSolution using np.linalg.solve:")
print(f"x = {X_solve[0]:.2f}, y = {X_solve[1]:.2f}, z = {X_solve[2]:.2f}")
```

# 10.2 Experiment: How Fast is Numpy?

```python
import numpy as np
import time

size = 1_000_000

# Setup data
list_a = list(range(size))
list_b = list(range(size))
np_a = np.arange(size, dtype=float)
np_b = np.arange(size, dtype=float)
```

## 1. Element-wise Addition:

```python
start = time.time()
result = [list_a[i] + list_b[i] for i in range(size)]
end = time.time()
print(f"   Python List time: {end - start:.5f} seconds")

start = time.time()
result = np_a + np_b
end = time.time()
print(f"   NumPy Array time: {end - start:.5f} seconds")
```

## 2. Element-wise Multiplication


```python
start = time.time()
result = [list_a[i] * list_b[i] for i in range(size)]
end = time.time()
print(f"   Python List time: {end - start:.5f} seconds")

start = time.time()
result = np_a * np_b
end = time.time()
print(f"   NumPy Array time: {end - start:.5f} seconds")
```

## 3. Dot Product


```python
start = time.time()
result = sum(list_a[i] * list_b[i] for i in range(size))
end = time.time()
print(f"   Python List time: {end - start:.5f} seconds")

start = time.time()
result = np.dot(np_a, np_b)
end = time.time()
print(f"   NumPy Array time: {end - start:.5f} seconds")
```

## 4. Matrix Multiplication


```python
mat_size = 1000
list_m1 = [[1.0] * mat_size for _ in range(mat_size)]
list_m2 = [[1.0] * mat_size for _ in range(mat_size)]
np_m1 = np.ones((mat_size, mat_size))
np_m2 = np.ones((mat_size, mat_size))

start = time.time()
result = [[sum(list_m1[i][k] * list_m2[k][j]
           for k in range(mat_size))
           for j in range(mat_size)]
           for i in range(mat_size)]
end = time.time()
print(f"   Python List time: {end - start:.5f} seconds")

start = time.time()
result = np.dot(np_m1, np_m2)
end = time.time()
print(f"   NumPy Array time: {end - start:.5f} seconds")

print("\n✓ Experiment complete! NumPy is significantly faster!")
```
